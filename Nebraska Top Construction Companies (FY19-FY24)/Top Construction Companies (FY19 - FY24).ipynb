```python
import pandas as pd   

contracts = pd.read_csv('Construction Contracts (FY19 - FY24).csv', low_memory=False)
```

```python
for i in contracts.columns:
    print(i)
```

```python
import pandasql as psql

query = """
SELECT 
    recipient_name,
    recipient_uei,
    naics_code,
    naics_description,
    recipient_country_name,
    recipient_address_line_1,
    recipient_city_name,
    recipient_state_name,
    recipient_zip_4_code,
    SUM(total_dollars_obligated) AS total,
    COUNT(*) AS count
FROM contracts
GROUP BY recipient_uei
ORDER BY total DESC
LIMIT 20
"""

construction_contracts = psql.sqldf(query, locals())
```

```python
construction_contracts['recipient_name'] = construction_contracts['recipient_name'].str.title()
construction_contracts['naics_description'] = construction_contracts['naics_description'].str.title()

def get_address(df):
    df['Address'] = f'{df['recipient_address_line_1']}, {df['recipient_city_name'].title()}, {df['recipient_state_name'].title()} {df['recipient_zip_4_code'][:5]}'

    return df

construction_contracts = construction_contracts.apply(get_address, axis=1)
construction_contracts = construction_contracts[['recipient_name', 'recipient_uei', 'total', 'count', 
                                                 'naics_code', 'naics_description', 'Address']]
```
